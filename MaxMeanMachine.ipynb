Получить среднее количество техники по их типу за календарный месяц (максимальное из среднего количества за первые 20 дней и среднего за последние 20 дней).

    data          - DataFrame, содержит первичный массив данных;
    data_f20_days - DataFrame/Series, содержит данные (количество по дням/среднее) за первые 20 дней месяца;
    data_l20_days - DataFrame/Series, содержит данные (количество по дням/среднее) за последние 20 дней месяца;
    data_max_mean - Series, содержит максимальное значение из 2 средних значений из data_f20_days и data_l20_days;
    fdata         - итоговый DataFrame, содержит тип техники и максимальное среднее.

```python
import pandas as pd
import json
```

```python
file = open('datasets/data.txt', 'r', encoding='utf-8')
data = file.read()
data = str(data).replace("'",'"')
data = json.loads(data)
```

```python
data = pd.json_normalize(data)
```

```python
# обработка NaN значений
data = data.fillna(0)
# замена типов данных
for i in data:
    if data[i].dtypes == 'float64':
        data[i] = data[i].astype('int64')
    else:
        continue
```

```python
# Фреймы с данными за первые 20 дней месяца и последние 20 дней месяца
data_f20_days = data.iloc[:, 5:25]
data_l20_days = data.iloc[:, -20:]

# Серии со средними значениями по каждой строке
data_f20_days = data_f20_days.mean(axis=1)
data_l20_days = data_l20_days.mean(axis=1)
```

```python
# конкатенация data_f20_days и data_l20_days по максимальному значению из 2 средних
data_max_mean = pd.concat([data_f20_days, data_l20_days], axis=1).max(axis=1)
```

```python
# слияние в итоговый DataFrame (столбец ParName с типами техники + максимальное среднее)
fdata = data.iloc[:, 1:2].merge(data_max_mean.to_frame(), left_index=True, right_index=True)
fdata = fdata.rename(columns={0: 'MaxMean'})
fdata['MaxMean'] = fdata['MaxMean'].astype('int64')
fdata
```
